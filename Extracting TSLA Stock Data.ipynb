### Plotting the Closing Price

```python
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['Close'], label='Close Price')
plt.title('Tesla Closing Price Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.legend()
plt.show()
```

### Plotting the Volume Traded

```python
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['Volume'], label='Volume Traded', color='orange')
plt.title('Tesla Trading Volume Over Time')
plt.xlabel('Date')
plt.ylabel('Volume Traded')
plt.legend()
plt.show()
```

### Calculating and Plotting Moving Averages

```python
data['20_SMA'] = data['Close'].rolling(window=20).mean()
data['50_SMA'] = data['Close'].rolling(window=50).mean()



plt.figure(figsize=(10, 6))
plt.plot(data.index, data['Close'], label='Close Price')
plt.plot(data.index, data['20_SMA'], label='20-Day SMA', color='green')
plt.plot(data.index, data['50_SMA'], label='50-Day SMA', color='red')
plt.title('Tesla Closing Price and Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()
```

### Calculating and Plotting Daily Returns

```python
data['Daily Return'] = data['Close'].pct_change()



plt.figure(figsize=(10, 6))
plt.plot(data.index, data['Daily Return'], label='Daily Return', color='purple')
plt.title('Tesla Daily Returns')
plt.xlabel('Date')
plt.ylabel('Daily Return')
plt.legend()
plt.show()
```

### Displaying Basic Statistics and Plotting Histogram

```python
print(data['Daily Return'].describe())


plt.figure(figsize=(10, 6))
data['Daily Return'].hist(bins=50, color='blue', alpha=0.7)
plt.title('Histogram of Tesla Daily Returns')
plt.xlabel('Daily Return')
plt.ylabel('Frequency')
plt.show()
```
